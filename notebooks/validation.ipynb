# helpers
___

```python
def get_deforestation_pixels_by_land_use_filename(year, biome):
    return CACHE_DIR / f"{LAND_USE_TYPE}/count/spatial_unit_deforestation_landuse_counts_b{biome}_y{year}.pkl"


def get_vegetation_pixels_by_land_use_filename(year, biome):
    return CACHE_DIR / f"{LAND_USE_TYPE}/count/spatial_unit_vegetation_landuse_counts_b{biome}_y{year}.pkl"


def get_annual_increase_in_deforestation_indicator_filename(first_year, last_year, biome):
    return CACHE_DIR / f"{LAND_USE_TYPE}/annual_increase_in_deforestation_indicator_b{biome}_from_y{first_year}_to_y{last_year}.pkl"


def get_accumulated_deforestation_indicator_filename(first_year, last_year, biome):
    return CACHE_DIR / f"{LAND_USE_TYPE}/accumulated_deforestation_indicator_b{biome}_from_y{first_year}_to_y{last_year}.pkl"


def get_vegetation_from_mask_filename(biome):
    return CACHE_DIR / f"{LAND_USE_TYPE}/vegetation_from_mask_b{biome}.pkl"


def get_vegetation_from_deforestation(first_year, last_year, biome):
    return CACHE_DIR / f"{LAND_USE_TYPE}/vegetation_from_deforestation_b{biome}_from_y{first_year}_to_y{last_year}.pkl"


def get_total_vegetation_filename(first_year, last_year, biome):
    return CACHE_DIR / f"{LAND_USE_TYPE}/vegetation_b{biome}_from_y{first_year}_to_y{last_year}.pkl"


def get_ratio_deforestation_to_vegetation_filename(first_year, last_year, biome):
    return CACHE_DIR / f"{LAND_USE_TYPE}/ratio_accumulated_deforestation_to_vegetation_indicator_b{biome}_from_y{first_year}_to_y{last_year}.pkl"


def load_dfr(filename, filters={}):
    assert filename.exists()
    
    dfr = pd.read_pickle(filename)

    mask = pd.Series(True, index=dfr.index)

    for col, value in filters.items():
        if isinstance(value, list):
            mask &= dfr[col].isin(value)
        else:
            mask &= dfr[col] == value

    return dfr[mask]
```

```python
filename = get_annual_increase_in_deforestation_indicator_filename(2000, 2025, biome)
assert filename.exists()
print(filename)

dfr = load_dfr(filename)
dfr
```

```python
reference_dict = {
    2000: reference_2000,
    2015: reference_2015,
    2024: reference_2024,
}
    
for year in [2000, 2015, 2024]:
    for spatial_unit in spatial_units:
        dfr = load_dfr(filename, filters={"spatial_unit": spatial_unit, "year": year})
        reference = reference_dict[year]
        total = dfr["deforestation_pixels"].sum()
        delta = total - reference
        percent = abs(delta) / reference * 100
        assert percent < 0.0025
        print(f"{spatial_unit}: year: {year}, num_pixels={total}, reference={reference}, delta={delta} perent={percent:.6f}")
```

# accumulated deforestation indicator (absolute counts)
___

```python
filename = get_annual_increase_in_deforestation_indicator_filename(2000, 2025, biome)
print(filename)
assert filename.exists()

years = np.arange(2000, 2025+1, 1).tolist()

annual_dfr = load_dfr(filename)

acc_dict = {}

for spatial_unit in spatial_units:
    _abs = 0
    acc_dict[spatial_unit] = {}
    for year in years:
        dfr = load_dfr(filename, filters={"spatial_unit": spatial_unit, "year": year})
        
        _rel = dfr["deforestation_pixels"].sum()
        _abs = _abs + _rel

        acc_dict[spatial_unit].update({
            year: {
                "relative": _rel,
                "absolute": _abs,
            }
        })
```

```python
vegetation_from_deforestation = {}
years_dict = {}
for year1 in years:
    total = 0
    years_dict[year1] = ""
    for year2 in years:
        if year2 <= year1:
            continue
        total = total + acc_dict["cs_5km"][year2]['relative']
        years_dict[year1] = years_dict[year1] + f",{year2}"
    vegetation_from_deforestation[year1] = total
```
